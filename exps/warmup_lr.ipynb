```python
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.optim import SGD
from models.resnet import ResNet
from models.blocks import Bottleneck, BasicBlock
```

```python
model = ResNet("cifar10", 'relu', True, Bottleneck, [2, 2, 2, 2])
optimizer = SGD(model.parameters(), lr=0.1 + 1e-8, momentum=0.9)
warmup_scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=200 , T_mult=1, eta_min=0, last_epoch=- 1)
```

```python
for i in range(200):
    warmup_scheduler.step()
    print(optimizer.param_groups[0]["lr"])
```
